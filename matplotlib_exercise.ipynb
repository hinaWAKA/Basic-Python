## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
# CSVファイルからデータを読み込む
hist_df = pd.read_csv(r"C:/Users/kupe20220413-1/python-assignment/Basic-Python/data/hist_data.csv")

# ヒストグラムの設定
plt.figure(figsize=(5, 3))

# aデータのヒストグラムを半透明のオレンジ色で描画
plt.hist(hist_df['aデータ'], bins=range(0, 56, 1), alpha=0.5, color='orange', label='aデータ')

# bデータのヒストグラムを半透明の水色で描画
plt.hist(hist_df['bデータ'], bins=range(0, 56, 1), alpha=0.5, color='cyan', label='bデータ')

# タイトル、軸ラベルを設定
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度 (℃)')
plt.ylabel('度数')

# 凡例を表示
plt.legend()

# 縦軸、横軸の範囲を設定
plt.xlim(0, 55)
plt.ylim(0, 27)

# ヒストグラムを表示
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
# CSVファイルからデータを読み込む
scatter_df = pd.read_csv(r"C:/Users/kupe20220413-1/python-assignment/Basic-Python/data/scatter_data.csv")

# クラスごとに色を指定（数値のクラスに対応するように変更）
colors = {0: 'red', 1: 'blue', 2: 'green'}

# 散布図を描画するためのプロット設定
plt.figure(figsize=(6, 4))

# クラスごとに散布図を描画
for class_value, color in colors.items():
    subset = scatter_df[scatter_df['class'] == class_value]
    plt.scatter(subset['length (cm)'], subset['width (cm)'], color=color, label=f'class{class_value}')

# 軸ラベルとタイトルの設定
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス0～2のlengthとwidthに関する散布図')

# グリッドと凡例の表示
plt.grid(True)
plt.legend()

# 散布図の表示
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
# CSVファイルからデータを読み込む
line_df = pd.read_csv(r"C:/Users/kupe20220413-1/python-assignment/Basic-Python/data/line_data.csv")

# 折れ線グラフの設定
plt.figure(figsize=(8, 5))

# 折れ線グラフを描画（青色で）
plt.plot(line_df['期間'], line_df['件数'], color='blue', marker='o')

# タイトル、軸ラベルを設定
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')

# 縦軸、横軸の範囲を設定
plt.ylim(101, 144)

# グラフにグリッドを追加
plt.grid(True)

# 横軸のラベルを傾けて表示
plt.xticks(rotation=45)

# 折れ線グラフを表示
plt.show()



# 棒グラフの設定（タイトルと軸ラベルを日本語で設定）
plt.figure(figsize=(8, 5))

# 棒グラフを描画（青色で）
plt.bar(line_df['期間'], line_df['費用'], color='blue')

# タイトル、軸ラベルを設定
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用(円)')

# 縦軸の範囲を設定（最大値を600000に変更）
plt.ylim(0, 599999)

# 横軸のラベルを傾けて表示
plt.xticks(rotation=45)

# 棒グラフにグリッドを追加
plt.grid(True)

# 棒グラフを表示
plt.show()
```
