## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
# CSVファイルからデータを読み込む
df = pd.read_csv(r"C:/Users/kupe20220413-1/python-assignment/Basic-Python/data/hist_data.csv")

# データのカテゴリ"aグループ"と"bグループ"の値を取得
a_group = df['aデータ']
b_group = df['bデータ']

# ヒストグラムをプロット
plt.hist([a_group, b_group], bins=range(10, 60, 5), label=['aデータ', 'bデータ'], color=['green', 'orange'], alpha=0.7)
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend(loc='upper right')
plt.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.4)
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
# CSVファイルからデータを読み込む
df = pd.read_csv(r"C:/Users/kupe20220413-1/python-assignment/Basic-Python/data/scatter_data.csv")

# クラスごとに色を指定
colors = {'class0': 'red', 'class1': 'blue', 'class2': 'green'}

# クラスごとに散布図を描画
for classname, color in colors.items():
    subset = df[df['class'] == classname]
    plt.scatter(subset['length (cm)'], subset['width (cm)'], color=color, label=classname)

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.grid(True)
plt.legend()
plt.show()
```
